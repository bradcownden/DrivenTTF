```python
import scipy.special
import numpy as np
```

```python
# Argument values
d = 4
Dpl = d
w = 0.25
```

```python
# True value of hypergeometric function
xrange = np.linspace(np.pi/2 * (0.9), np.pi/2 *(1.1), 1000)
(np.cos(xrange)) ** Dpl * scipy.special.hyp2f1((Dpl + w)/2, (Dpl - w)/2, d/2,
                     np.sin(xrange) * np.sin(xrange))
```
